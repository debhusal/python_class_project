## Removing the null data point

Now let us check if our data contains null value or not. To analyze the data, we will need to normalize the data. Mostly for our experiment in our lab, we usually pre-fill 1/5 of the minimum value in the row in the empty cell. That way the missing value will be very small to make a impact on the end result, and another good thing is that we will not need to worry about the empty data.

Let's check how many non-null data we have out of 5196 number of cells in each column. We can check that using the following code.

```python
data.info() # Checking number of non-null data in each column
```

We can see above that the data type of the dataframe is object, so, let us first convert the type of dataframe to numeric.

```python
df1 = data.apply(pd.to_numeric)
```

Now, let us normalize the data and fill the missing values as the 1/5th of the minimum value in the row.

```python
data_df = df1.copy()
new_df = data_df.apply(lambda row: row.fillna(row.min()/5), axis=1)
```

Let us confirm that we have filled the missing values in the dataframe.

```python
new_df.head()
```

We can confirm that we have success fully updated the missing values in the dataframe. Now it will be easier for us to anayze the data.

Now since we have successfully filled the missing values, let us save the data in the csv file. Let's save the data in the csv file.


```python
new_df.to_csv("./results/data_filled_missing_value.csv")
```


# Analysis of the Data

Now to do the analysis of the data, let's begin with plotting the heatmap of our intensities. The x-axis of our heatmap will be different single cells which analyzed. The y-axis of our heat map will be the `Mean m/z` value. And the color will be the intentsity for `m/z` for each cell.

```python
sns_plot = sns.heatmap(new_df, annot=False, cmap='coolwarm', cbar_kws={'label': 'Intensities'})
sns_plot.set(xlabel='Single Cells')
sns_plot.get_figure().savefig("./results/heatmap_overall_initial.png") 
```

Looking at the heat map, we found out that most of the data has intensity of less than 5 because most of the color of the data looks dark blue. This doesn't give us good information we would first like to check how the intensity is distributed across `Mean m/z` value in a single cell. To check that let us plot a scatter plot of `m/z` value for cell `48hQ_3`.

```python
scatter_df = new_df.copy()
ax = scatter_df.reset_index().plot(kind='scatter', x='Mean m/z', y='48hQ_3', s=1)
ax.set(ylabel="Intensity")
ax.set(xlim=(200, 1500))
ax.set(title="Intensity of 48hQ_3 Cell across m/z values")
ax.get_figure().savefig("./results/scatter_plot_48hQ_3.png") 
```

Looking at the scatter plot we can see that we have some very high intensity but most of the intensities seemsto be closer to 0 value. 

## Log transformation of data

Because intensities look distributed across very high range, to make the heat map more clear, we would like to log base 10 transform the dataframe. This will bring all the data below 6 as the maximum value in the cell seems to be around 60,000.

```python
numeric_df = new_df.apply(lambda x: np.log10(x))
numeric_df.head()
```

Now we can see above that we are seeing negative values in the data. This means that log transformation of our dataframe was a success. All the intensities which were less than 1 in value have been transformed to negative value. And all the intensities which were larger than 1 in value have been transformed to positive value.

Let's confirm that the data have been successfully transformed by plotting the scatter plot one more time.

```python
scatter_df = numeric_df.copy()
ax = scatter_df.reset_index().plot(kind='scatter', x='Mean m/z', y='48hQ_3', s=1)
ax.set(ylabel="Log_10(intensity)")
ax.set(xlim=(200, 1500))
ax.set(title="Log transformed intensity of 48hQ_3 Cell across m/z values")
ax.get_figure().savefig("./results/scatter_plot_48hQ_3_log_transformed.png") 
```

We have successfully log transformed the intensity in the dataframe.

## Heatmap of intensity

Now, we will go ahead and try to plot the heatmap of the log transformed heatmap and see what would be the result like. 

```python
sns_plot = sns.heatmap(numeric_df, annot=False, cmap='coolwarm', vmax=3, vmin=-3, cbar_kws={'label': 'Log transformed Intensities'})
sns_plot.set(xlabel='Single Cells')
sns_plot.get_figure().savefig("./results/heatmap_overall_final.png") 
```

We can see in the above image that the heatmap is more robust this time. Outliers hasn't impacted the heatmap as it did last time. We can see for some `m/z` value in the top of the heatmap, the intensity is larger than 0. This means that the intensity was larger than 1 in our actual data. So, the metabloids of that mass were present in all of the cells irrespective of the quenching duration.
